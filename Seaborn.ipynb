# scatter plot  -> fig level function 
hue,style,size are parameters to categorise data based on certain parameter and provides a legend
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='sex',style='time',size='sex')

**fig level functions are more general in kind bcs if we plot a line then it has its own function but if we use fig function that is ;rel 'function', then
we can use it for both scatter , line and many more**


```python
# fig level function
# we can use 'hue' parameter to categorise the data and automatically a legend gets added
sns.relplot(data=tips, x='total_bill',y='tip', kind='scatter',hue='sex',style='time',size='size')
```

# facet plot

```python
# It will work only with fig level functions such as relplot
# So we cannot do categorial plot in axis -level plot functions such as sns.scatterplot() as it doesnot have access to the whote figure,
# we have to show the plot with categories together only
sns.scatterplot(data=tips, x='total_bill',y='tip',col='sex') 
```

# Kde Plot
**It is used to draw a Gaussin Kernel plot where in y we have the density or probability of occurance of any point in x axis**

```python
sns.kdeplot(data=tips,x='tip',hue='sex') # col doesnot work for axis level plots as they cannot use entire fig to split on multiple
```

```python
sns.displot(data=tips,x='tip',kind='kde',hue='sex',col='sex')
```

```python
sns.displot(data=tips,x='tip',kind='kde',hue='sex',fill=True)
```

```python
# Bivariate histograms
#we can use both x and y values to create 2d tile to show relation between columns 
sns.displot(kind='hist',data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# the more is the dark color, more is the density / probability of giving that amount of tip with that much of ttl bill
sns.displot(data=tips,x='total_bill',y='tip',kind='kde',fill=True)
```

# Matrix Plots
1. Heatmap
2. Clustermap

***pivot => used to convert long data into wide data. similar to confusion metrix***

```python
gap_=gap.pivot(index='year',columns='country',values='lifeExp')
```

```python
gap_
```

```python
# heatmap  -> it only has axis level function , not fig level
plt.figure(figsize=(10,5))
sns.heatmap(gap_)
```

```python
# plotting for single contitent
plt.figure(figsize=(10,5))
sns.heatmap(gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp'),annot=True)
```

```python
sns.heatmap(gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp'),annot=True,cmap='summer',linewidth=1)
# cmap is colour map
```
