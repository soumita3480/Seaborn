# Kde Plot
**It is used to draw a Gaussin Kernel plot where in y we have the density or probability of occurance of any point in x axis**

```python
sns.kdeplot(data=tips,x='tip',hue='sex') # col doesnot work for axis level plots as they cannot use entire fig to split on multiple
```

```python
sns.displot(data=tips,x='tip',kind='kde',hue='sex',col='sex')
```

```python
sns.displot(data=tips,x='tip',kind='kde',hue='sex',fill=True)
```

```python
# Bivariate histograms
#we can use both x and y values to create 2d tile to show relation between columns 
sns.displot(kind='hist',data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# the more is the dark color, more is the density / probability of giving that amount of tip with that much of ttl bill
sns.displot(data=tips,x='total_bill',y='tip',kind='kde',fill=True)
```

## Categorical Scatter Plot (it talks about relation between 2 axis)

**Swarm PLot**

```python
# Strip Plot  (Same as Scatter Plot, it takes one cate col and one numeric col and plots the data)
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
sns.get_dataset_names()
```

```python
tips
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='day')
```

```python
#Ideally the strip plot should look like this but for better readibility seaborn adds noise and scatters the points, we can remove that using jitter fun
sns.stripplot(data=tips,x='day',y='total_bill',hue='day',jitter=False) # axis level function
```

```python
# fig level function for categorical plot
sns.catplot(data=tips,x='day',y='tip',kind='strip',hue='day')
```

```python
# Swarmplot
# Swarm plot is very similar to stripplot. only is runs a inernal algorithm which also shows the distribution or spread of y axis value for a certain x value
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

## Categorical Distribution PLot (It tells about dis of a single col value)
**Box Plot**
**Violin Plot**


### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box Plot
sns.boxplot(data=tips,y='total_bill')
```

```python
# plot based on category (axis level plot)
sns.boxplot(data=tips,x='sex',y='total_bill',hue='sex')
```

```python
# box plot for fig level
sns.catplot(data=tips,y='total_bill',hue='sex',kind='box')
```

```python
sns.boxplot(data=tips,y='total_bill',hue='sex',x='day')
```

```python
# violin plot (box plot+ kde plot)
sns.violinplot(data=tips,y='total_bill',hue='sex')
```

```python
sns.catplot(data=tips,y='total_bill',hue='sex',kind='violin')
```
