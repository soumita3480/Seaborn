```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
# scatter plot  -> fig level function 
# hue,style,size are parameters to categorise data based on certain parameter and provides a legend
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='sex',style='time',size='sex')
```

**fig level functions are more general in kind bcs if we plot a line then it has its own function but if we use fig function that is ;rel 'function', then
we can use it for both scatter , line and many more**


```python
# fig level function
# we can use 'hue' parameter to categorise the data and automatically a legend gets added
sns.relplot(data=tips, x='total_bill',y='tip', kind='scatter',hue='sex',style='time',size='size')
```

```python
gap=px.data.gapminder()
ind=gap[gap['country']=='India']
ind
```

```python
# usually we use lineplot to plot data based on time such as day, month, year
sns.lineplot(data=ind, x='year',y='lifeExp')
```

```python
sns.relplot(data=tips, x='total_bill',y='tip',kind='line')
```

```python
ind_pak_cin=gap[gap['country'].isin(['India','China','Pakistan'])]
ind_pak_cin
```

```python
sns.relplot(data=ind_pak_cin,x='year',y='lifeExp',hue='country',kind='line')
```

```python
ind_bra_ger=gap[gap['country'].isin(['India','Brazil','Germany'])]
ind_bra_ger
```

```python
sns.relplot(data=ind_bra_ger,x='year',y='lifeExp',kind='line',style='continent')
```
